# Blutzuckerübersteigung bei männlichen Beobachteten

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
heartData = pd.read_csv("heart-failure-prediction/heart.csv")
heartData.head()
```

```python
#Cholesterin Werte von Herzkreislauf Erkrankten auslesen
heartData = heartData[(heartData["Cholesterol"]!=0) & (heartData["Sex"]!="F")]
fastingBSColumn = heartData["FastingBS"]
fastingBS = fastingBSColumn.value_counts().sort_index()
fastingBSCounts = fastingBSColumn.drop_duplicates().sort_values()
```

```python
#Balkendiagramm zu bleibenden Beobachtungen erstellen
plt.bar(fastingBSCounts,fastingBS,color="skyblue")
plt.title("Blutzucker>120 von mänlichen Beobachteten")
plt.xlabel("Blutzucker")
plt.ylabel("Anzahl")
plt.grid(axis="y")
plt.show()
```

```python
#Balkendiagramm zu bleibenden Beobachtungen erstellen
unhealthyData = heartData[heartData["HeartDisease"]==1]

fastingBSColumn = unhealthyData["FastingBS"]
fastingBS = fastingBSColumn.value_counts().sort_index()
fastingBSCounts = fastingBSColumn.drop_duplicates().sort_values()

plt.bar(fastingBSCounts,fastingBS,color="skyblue")
plt.title("Blutzucker>120 von mänlichen herzkreislauferkrankten")
plt.xlabel("Blutzucker")
plt.ylabel("Anzahl")
plt.grid(axis="y")
plt.show()
```

```python
#Balkendiagramm zu bleibenden Beobachtungen erstellen
healthyData = heartData[heartData["HeartDisease"]==0]

fastingBSColumn = healthyData["FastingBS"]
fastingBS = fastingBSColumn.value_counts().sort_index()
fastingBSCounts = fastingBSColumn.drop_duplicates().sort_values()

plt.bar(fastingBSCounts,fastingBS,color="skyblue")
plt.title("Blutzucker>120 von mänlichen gesunden Patienten")
plt.xlabel("Blutzucker")
plt.ylabel("Anzahl")
plt.grid(axis="y")
plt.show()
```
