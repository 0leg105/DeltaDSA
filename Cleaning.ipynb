# Entfernen von zu hohen Abweichungen zur Steigerung des Korrelationskoeffizienten

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
heartData = pd.read_csv("heart-failure-prediction/heart.csv")
heartData.head()
```

## Entfernung von fehlenden Cholesterin-Beobachtungen und Frauen

```python
heartData = heartData[(heartData["Cholesterol"] != 0) & (heartData["Sex"] == "F")]
```

```python
plt.title("Zusammenhang zwischen Cholesterin und Ruheblutdruck bei mnnlichen Patienten mit Herzerkrankungen")
plt.scatter(heartData["Cholesterol"], heartData["RestingBP"])
m, b = np.polyfit(heartData["Cholesterol"], heartData["RestingBP"], 1)
plt.plot(heartData["Cholesterol"], m*heartData["Cholesterol"] + b, color='red', label='Trendlinie')
plt.xlabel("Cholesterol")
plt.ylabel("Ruheblutdruck")
plt.legend()
plt.show()

correlation = heartData['Cholesterol'].corr(heartData['RestingBP'])
print("Korrelationskoeffizient: " + str(correlation))
```
